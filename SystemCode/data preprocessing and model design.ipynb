```python
data = pd.read_csv('./Coursera_courses.csv', index_col=0)
```

```python
# Translations of more than 5,000 characters need to be translated in two separate parts
def translation(texts):
    from googletrans import Translator
    translator = Translator(service_urls=['translate.google.com'])
    if len(texts) > 5000:
        seperate_index = 0
        for i in range(4999, 0, -1):
            if texts[i] == ',':
                text_1 = texts[:i]
                text_2 = texts[i:]
                texts = text_1 + '(split)' + text_2
                break
        texts_array = texts.split('(split)')
        translated_texts = ''
        for i in range(len(texts_array)):
            translated_texts_once = translator.translate(texts_array[i]).text
            print(translated_texts_once)
            print('*' * 100)
            translated_texts += (translated_texts_once + ' ')
    else:
        translated_texts = translator.translate(texts).text
    return translated_texts
```

```python
# 去除首尾空格
def trim(s):
    if s[0] == ' ':
        return trim(s[1:])
    elif s[-1] == ' ':
        return trim(s[:-1])
    else:
        return s
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support import ui
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from time import sleep

wait_time = 180

url = 'https://www.coursera.org/search?query={}&'
data_test = data.copy()
data_test.insert(loc=8, column='skills', value='None')

driver = webdriver.Chrome(r'F:\project1\chromedriver.exe')
driver.maximize_window()
driver.implicitly_wait(2)

for i in range(len(data_test)):
    print(i)
    course_name = data_test.loc[i, 'course_name']
    if '&' in course_name:
        course_name = course_name.replace('&', 'and')
    driver.get(url.format(course_name))
    
    element = driver.find_element(By.CSS_SELECTOR, 'h1 > div > span')
    if '未找到关于' in element.text:
        driver.back()
        continue
        
    course_name_page = driver.find_element(By.CSS_SELECTOR, '.css-18msmec > li:nth-child(1) h2')
    if trim(course_name) == trim(course_name_page.text):
        elements = driver.find_elements(By.CSS_SELECTOR, '.css-18msmec > li:nth-child(1)  p')
        have_skills = '您将获得的技能'
        if len(elements) >= 3:
            for j in range(len(elements)):
                if have_skills in elements[j].text:
                    skills = elements[j].text.replace(have_skills + '： ', '')
                    print(skills)
                    data_test.loc[i, 'skills'] = skills
                    continue
        
    ac = ActionChains(driver)
    ac.move_to_element(
        driver.find_element(By.CSS_SELECTOR, '.css-18msmec > li:nth-child(1)')
    ).click().perform()
    
    if len(driver.window_handles) == 1:         # If you cannot find the corresponding course
        driver.back()
        continue
          
    sleep(1)
    driver.switch_to.window(driver.window_handles[-1])
    elements = driver.find_elements(By.CSS_SELECTOR, '.css-uqope5 > li')
    
    if len(elements) != 0:
        skills = ''
        for element in elements:
            skills += (element.text + ',')
        data_test.loc[i, 'skills'] = skills
        driver.close()      # Close the current tab
        driver.switch_to.window(driver.window_handles[0])
    else:
        driver.close()      # Close the current tab
        driver.switch_to.window(driver.window_handles[0])
sleep(3)
driver.quit()
```

# Recommender Algorithm

## User_preference courses recommender

```python
data_test = data_4.copy()
```

```python
"""
    user inputs：
        University
        Course Level
        Course Language
        Preferred Skills 
"""
user_university = 'Coursera Project Network'
user_level = 'Beginner'
user_language = 'English'
user_skills = 'Marketing, Strategy, Content, Machine Learning'

similarity_list = np.zeros(len(data_test))
for i in range(len(data_test)):
    if data_test.loc[i, 'university_name'] == user_university:
        similarity_list[i] += 0.1
    if data_test.loc[i, 'course_language'] == user_language:
        similarity_list[i] += 0.3
    if data_test.loc[i, 'course_level'] == 'Beginner':
        if user_level == 'Beginner':
            similarity_list[i] += 0.1
        elif user_level == 'Intermediate':
            similarity_list[i] += 0.05
    if data_test.loc[i, 'course_level'] == 'Intermediate':
        if user_level == 'Beginner':
            similarity_list[i] += 0.05
        elif user_level == 'Intermediate':
            similarity_list[i] += 0.1
        elif user_level == 'Advanced':
            similarity_list[i] += 0.05
    if data_test.loc[i, 'course_level'] == 'Advanced':
        if user_level == 'Intermediate':
            similarity_list[i] += 0.05
        elif user_level == 'Advanced':
            similarity_list[i] += 0.1
    rating = data_test.loc[i, 'course_rating']
    if rating == 'Not Calibrated':
        similarity_list[i] += 0
    else:
        similarity_list[i] += (float(rating) - 4) * 0.1
    print(similarity_list[i])
```

```python
from sentence_transformers import SentenceTransformer, util
import torch
from tqdm import tqdm

embedder = SentenceTransformer('all-MiniLM-L6-v2')

# Corpus with example sentences
corpus = list(data_test['skills'])
corpus_embeddings = embedder.encode(corpus, convert_to_tensor=True)

query = user_skills
query_embedding = embedder.encode(query, convert_to_tensor=True)
cos_scores = util.cos_sim(query_embedding, corpus_embeddings)[0]
cos_scores = cos_scores.cpu().detach().numpy()
similarity_list += cos_scores
```

## Subsequent content-based user-courses recommender

```python
from sentence_transformers import SentenceTransformer, util
import torch
from tqdm import tqdm

embedder = SentenceTransformer('all-MiniLM-L6-v2')

# Corpus with example sentences
corpus = list(data_test['course_detail'])
corpus_embeddings = embedder.encode(corpus, convert_to_tensor=True)

# Query sentences:
for i in tqdm(range(len(data_test))):
    query_index = i 
    query = data_test.loc[query_index, 'course_detail']
    query_embedding = embedder.encode(query, convert_to_tensor=True)
    cos_scores = util.cos_sim(query_embedding, corpus_embeddings)[0]
    cos_scores = cos_scores.cpu().detach().numpy()
    if i == 0:
        similarity_matrix = [cos_scores]
    else:
        similarity_matrix.append(cos_scores)
```

```python
# Calculate the similarity between all courses and save it as a matrix
# Just read the similarity each time
# After reading the similarity between courses, add the corresponding values for university_name, 
# course_level, course_language and course_rating to the recommendation

university_matrix = np.zeros((data_test.shape[0], data_test.shape[0]))
language_matrix = np.zeros((data_test.shape[0], data_test.shape[0]))
level_matrix = np.zeros((data_test.shape[0], data_test.shape[0]))

for i in tqdm(range(len(similarity_matrix))):
    for j in range(len(similarity_matrix)): 
        if data_test.loc[i, 'university_name'] == data_test.loc[j, 'university_name']:
            university_matrix[i][j] = 0.1
```

```python
for i in tqdm(range(len(similarity_matrix))):
    for j in range(len(similarity_matrix)): 
        if data_test.loc[i, 'course_language'] == data_test.loc[j, 'course_language']:
            language_matrix[i][j] = 0.3
```

```python
for i in range(len(data_test)):
    if data_test.loc[i, 'course_level'] == 'Beginner-friendly':
        data_test.loc[i, 'course_level'] = 'Beginner'
    if data_test.loc[i, 'course_level'] == 'Conversant':
        data_test.loc[i, 'course_level'] = 'Advanced'
```

```python
for i in tqdm(range(len(similarity_matrix))):
    for j in range(len(similarity_matrix)): 
        if data_test.loc[i, 'course_level'] == 'Beginner':
            if data_test.loc[j, 'course_level'] == 'Beginner':
                level_matrix[i][j] = 0.1
            elif data_test.loc[j, 'course_level'] == 'Intermediate':
                level_matrix[i][j] = 0.05
        if data_test.loc[i, 'course_level'] == 'Intermediate':
            if data_test.loc[j, 'course_level'] == 'Beginner':
                level_matrix[i][j] = 0.05
            elif data_test.loc[j, 'course_level'] == 'Intermediate':
                level_matrix[i][j] = 0.1
            elif data_test.loc[j, 'course_level'] == 'Advanced':
                level_matrix[i][j] = 0.05
        if data_test.loc[i, 'course_level'] == 'Advanced':
            if data_test.loc[j, 'course_level'] == 'Advanced':
                level_matrix[i][j] = 0.1
            elif data_test.loc[j, 'course_level'] == 'Intermediate':
                level_matrix[i][j] = 0.05
```

```python
rating_matrix = np.zeros((data_test.shape[0], data_test.shape[0]))
for i in tqdm(range(len(data_test))):
    for j in range(len(data_test)):
        rating = data_test.loc[j, 'course_rating']
        if rating == 'Not Calibrated':
            rating_matrix[i][j] = 0
        else:
            rating_matrix[i][j] = (float(rating) - 4) * 0.1
```

```python
final_matrix_detail = similarity_matrix + level_matrix + university_matrix + language_matrix + rating_matrix
```
